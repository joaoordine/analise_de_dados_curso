## Processing raw data from Fragpipe

```python
# Download raw table from fragpipe 
## I manually removed additional replicates in compartiments that have more than 2 (some were lost)
```

```python
import pandas as pd

# Load the file into a DataFrame
df = pd.read_csv('data/MCF7_EGF_PROTEINS_COMBINED.tsv', sep='\t')

# Replace spaces with underscores in column names
df.columns = df.columns.str.replace(' ', '_')

# Display the first few rows to verify the changes
df.head()
```

```python
# Identify columns that are numerical (or should be numeric)
df_filtered = df
for col in df_filtered.columns:
    # Replace commas with dots and convert to numeric, errors='coerce' handles non-numeric values gracefully
    df_filtered[col] = df_filtered[col].astype(str).str.replace(',', '.').astype(float, errors='ignore')

# Display the first few rows to verify the changes
df_filtered.head()
```

```python
# Apply the filters on df_filtered
df_filtered2 = df_filtered[(df_filtered['Combined_Total_Spectral_Count'] > 3) & 
                           (df_filtered['Protein_Probability'] > 0.7)]

# Display the first few rows to verify the result
df_filtered2.head()
```

```python
# Normalization - relative abundance
## Make a copy of df_filtered2 to apply normalization
df_filtered2_norm = df_filtered2.copy()

## Identify columns ending with '_MaxLFQIntensity'
intensity_columns = [col for col in df_filtered2_norm.columns if col.endswith('_MaxLFQIntensity')]
intensity_columns
```

```python
## Normalize each '_MaxLFQIntensity' column in the new DataFrame
for col in intensity_columns:
    col_sum = df_filtered2_norm[col].sum()  # Sum of the column
    df_filtered2_norm[col] = (df_filtered2_norm[col] / col_sum) * 100  # Relative abundance calculation

## Display the first few rows to verify the result
df_filtered2_norm[intensity_columns].head()
```

```python
import os

## Specify the directory and file path
output_dir = "./data"
output_file = "df_filtered2_norm.tsv"

## Save the DataFrame as a TSV file
df_filtered2_norm.to_csv(os.path.join(output_dir, output_file), sep='\t', index=False)
```
