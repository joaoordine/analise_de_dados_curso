# Proposta de projeto desenvolvido em sala de aula

Fazer um programa para executar:
- Grupo 1
  - Obter uma sequência de nucleotídeos automaticamente do NCBI dado o ID, [exemplo](https://www.ncbi.nlm.nih.gov/nuccore/186972394);
- Grupo 2
 - Calcular o reverso da sequência;
 - Calcular o reverso complementar da sequência;
 - Fazer a transcrição;
 - Fazer a tradução;
- Grupo 3
 - Obter os [kmers](https://en.wikipedia.org/wiki/K-mer) da sequência;
 - Fazer um blast da sequência contra o banco de nucleotídeos do NCBI.

Fazer repositório do github, e fazer *fork* para todos os membros do grupo. Reunir programas e fazer aplicação linha de comando seguindo padrão de projeto exemplo do curso.
